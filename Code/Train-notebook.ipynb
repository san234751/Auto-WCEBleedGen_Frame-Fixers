```python
import os
from ultralytics import YOLO
from google.colab.patches import cv2_imshow
import cv2
import pandas as pd

def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,filename))
        if img is not None:
            images.append([img,filename])
    return images

imgs=load_images_from_folder('/content/drive/MyDrive/train_data/Test/Test_2')

model_path='/content/drive/MyDrive/train_data/runs/detect/train3/weights/best.pt'

# Load a model
model = YOLO(model_path)  # load a custom model
array=[]

threshold = 0.5

for image in imgs:

    results = model(image[0])[0]

    for result in results.boxes.data.tolist():
        x1, y1, x2, y2, score, class_id = result

        cv2.rectangle(image[0], (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
        if y1-10>10:
          cv2.putText(image[0], results.names[int(class_id)].upper()+' '+str(round(score,3)), (int(x1), int(y1 - 10)),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 1)
        else:
          cv2.putText(image[0], results.names[int(class_id)].upper()+' '+str(round(score,3)), (int(x1), int(y2 + 15)),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 1)

    # cv2_imshow(image[0])
    midarray=[]
    if len(results.boxes.data.tolist())<1:
      midarray=[image[1],'Non-Bleeding']
    else:
      midarray=[image[1],'Bleeding']
    array.append(midarray)
    # cv2.imwrite(f"/content/drive/MyDrive/train_data/Detection/Test_2/{image[1]}",image[0])
    cv2.waitKey(0)
array.sort()
predictions_test_dataset_1=pd.DataFrame(array,columns=['Image Name', 'Predicted Class label'])
predictions_test_dataset_1.to_csv('predictions_test_dataset_1.csv', index=False)
```
